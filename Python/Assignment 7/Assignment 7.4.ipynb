##### Dmart file = df

```python
df = pd.read_csv(path)
```

```python
df['Date'] = pd.to_datetime(df['Date/Time'])
df['Month'] = df['Date'].dt.month
df['Week'] = df['Date'].dt.week
```

```python
df_open = pd.DataFrame()
df_open = df.groupby(['Month','Week'])['Open'].first()
#display(df_open)

df_close = pd.DataFrame()
df_close = df.groupby(['Month','Week'])['Close'].last()
#display(df_close)

#newdf = df_open.join([df_close])
newdf = pd.concat([df_open,df_close],axis=1)
newdf['ROC'] = ((newdf['Close']/newdf['Open'])-1)*100
#display(newdf)

Returns = pd.DataFrame()
Returns['ROC'] = newdf['ROC']
display(Returns)
```
