```python
df = pd.read_csv(path)
```

```python
(df['ROC']) = ((df['Close'] - df['Open']) / df['Open'])*100
```

```python
ROC_positive = df[df['ROC']>0.5]
display(ROC_positive)
```

```python
Positive_sum = ROC_positive['ROC'].sum()
print('The total ROC of the ROC > 0.5% dataframe is: ',Positive_sum)

Positive_count = ROC_positive['ROC'].count()
print('The total number of trades in ROC > 0.5% dataframe is: ',Positive_count)
```

```python
ROC_negative = df[df['ROC']<-0.5]
display(ROC_negative)
```

```python
Negative_sum = ROC_negative['ROC'].sum()
print('The total ROC of the ROC < -0.5% dataframe is: ',Negative_sum)

Negative_count = ROC_negative['ROC'].count()
print('The total number of trades in ROC < -0.5% dataframe is: ',Negative_count)
```
