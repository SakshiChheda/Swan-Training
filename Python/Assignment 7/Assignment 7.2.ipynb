```python
df = pd.read_csv(path)
```

```python
(df['ROC']) = ((df['Close'] - df['Open']) / df['Open'])*100
```

```python
df['Type_of_Trade'] = df['ROC'].apply (lambda x: 1 if x > 0 else 0)
```

```python
Winners = df[df['Type_of_Trade']==1]
display(Winners)
```

```python
TotalROCWinners = Winners['ROC'].sum()
print('The total ROC of winners is: ',TotalROCWinners)
```

```python
Losers = df[df['Type_of_Trade']==0]
display(Losers)
```

```python
TotalROCLosers = Losers['ROC'].sum()
print('The total ROC of losers is: ',TotalROCLosers)
```
