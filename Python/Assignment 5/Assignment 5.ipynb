```python
import urllib
import pandas as pd
import numpy as np
import datetime
import zipfile, os
```

```python
DownloadStart = input('You would like to download bhavcopies from which date? (please input in dd/mm/yyyy format)\n')
DownloadStop = input('You would like to download bhavcopies till which date? (please input in dd/mm/yyyy format)\n')
```

```python
#getting the list of dates

from datetime import datetime,timedelta 
StartDownload = datetime.strptime(DownloadStart,'%d/%m/%Y')
StopDownload = datetime.strptime(DownloadStop,'%d/%m/%Y')

DateIndex = pd.date_range(StartDownload,StopDownload-timedelta(days=0),freq='d')
print(DateIndex)
```

```python
#downloading and unzipping the files in one folder

import requests
from zipfile import ZipFile
import os.path
    
link1 = "https://www1.nseindia.com/archives/fo/bhav/fo"
link2 = ".zip"
filenamehelp = "\\fo"
path = r"C:\Users\admin\Desktop\BHAVCOPIES"


i = 0
while i < len(DateIndex):
    for i in range(0,len(DateIndex)):
        Year = DateIndex[i].strftime('%y')
        Month = DateIndex[i].strftime('%m')
        Day = DateIndex[i].strftime('%d')
        finallink = link1 + Day + Month + Year + link2
        r = requests.get(finallink)
        filename = filenamehelp + Day + Month + Year + link2
        completename = path + filename
        with open(completename,'wb') as f:
            f.write(r.content)
        with ZipFile(completename,'r') as zip:
            zip.printdir()
            zip.extractall(path)
        i+=1
```
